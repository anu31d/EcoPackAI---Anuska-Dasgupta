```python
np.random.seed(42)
df["predicted_cost"] = np.random.uniform(5, 50, len(df))
df["predicted_co2"] = np.random.uniform(0.1, 5, len(df))
```

```python
ranking_config = {
    "weights": {
        "cost": 0.3,
        "co2": 0.4,
        "suitability": 0.3
    },
    "constraints": {
        "min_recyclability": "B",
        "max_cost": 40,
        "min_suitability": 50
    },
    "top_n": 3
}
```

```python
config_path = f"{CONFIG_DIR}/ranking_weights.yaml"
with open(config_path, "w") as f:
    yaml.dump(ranking_config, f)

config_path
```

```python
def min_max(series, reverse=False):
    norm = (series - series.min()) / (series.max() - series.min())
    return 1 - norm if reverse else norm

df["cost_norm"] = min_max(df["predicted_cost"], reverse=True)
df["co2_norm"] = min_max(df["predicted_co2"], reverse=True)
df["suit_norm"] = min_max(df["Material_Suitability_Score"])
```

```python
w = ranking_config["weights"]

df["ranking_score"] = (
    w["cost"] * df["cost_norm"] +
    w["co2"] * df["co2_norm"] +
    w["suitability"] * df["suit_norm"]
)
```

```python
recy_map = {"A":4, "B":3, "C":2, "D":1}
min_recy = recy_map[ranking_config["constraints"]["min_recyclability"]]

df = df[
    (df["predicted_cost"] <= ranking_config["constraints"]["max_cost"]) &
    (df["Material_Suitability_Score"] >= ranking_config["constraints"]["min_suitability"]) &
    (df["Recyclability Category"].map(recy_map) >= min_recy)
]
```

```python
df["rank"] = df.groupby("Material ID")["ranking_score"]\
                .rank(ascending=False, method="dense")

top_n = ranking_config["top_n"]

ranked = df[df["rank"] <= top_n]\
            .sort_values(["Material ID","rank"])
```

```python
ranker_code = '''
import pandas as pd
import yaml

def rank_materials(df, config_path):
    with open(config_path) as f:
        cfg = yaml.safe_load(f)

    w = cfg["weights"]
    c = cfg["constraints"]
    top_n = cfg["top_n"]

    def min_max(series, reverse=False):
        norm = (series - series.min()) / (series.max() - series.min())
        return 1 - norm if reverse else norm

    df = df.copy()
    df["cost_norm"] = min_max(df["predicted_cost"], reverse=True)
    df["co2_norm"] = min_max(df["predicted_co2"], reverse=True)
    df["suit_norm"] = min_max(df["Material_Suitability_Score"])

    df["ranking_score"] = (
        w["cost"] * df["cost_norm"] +
        w["co2"] * df["co2_norm"] +
        w["suitability"] * df["suit_norm"]
    )

    recy_map = {"A":4,"B":3,"C":2,"D":1}
    min_recy = recy_map[c["min_recyclability"]]

    df = df[
        (df["predicted_cost"] <= c["max_cost"]) &
        (df["Material_Suitability_Score"] >= c["min_suitability"]) &
        (df["Recyclability_Category"].map(recy_map) >= min_recy)
    ]

    df["rank"] = df.groupby("product_id")["ranking_score"]\
                    .rank(ascending=False, method="dense")

    return df[df["rank"] <= top_n].sort_values(["product_id","rank"])
'''
```

```python
ranker_path = f"{SRC_DIR}/ranker.py"
with open(ranker_path, "w") as f:
    f.write(ranker_code)

ranker_path
```
