```python
import pandas as pd
import numpy as np
from pathlib import Path

# Load
raw_fp = RAW_PATH
df = pd.read_csv(raw_fp)
print("Loaded dataframe shape:", df.shape)
df.head(3)
```

```python
# Basic info
df.info()

# Missing summary
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing>0]
print("Columns with missing values:\n", missing)

# Save a quick CSV of missing counts
missing.to_csv(os.path.join(DOCS_DIR, "missing_value_counts_raw.csv"))
```

```python
df_raw = df.copy()
df = df_raw.copy()
```

```python
# Auto-detect types
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

likely_missing_cols = [
    "Biodegradation Time (days)",
    "Recyclability %",
    "Waste Reduction Impact %",
    "Carbon Footprint",
    "Supplier Region",
    "Suitable Product Categories"
]

# Show what exists
print("Numerical columns:", num_cols)
print("Categorical columns:", cat_cols)
print("Likely missing columns present in dataset:", [c for c in likely_missing_cols if c in df.columns])
```

```python
# Replace common bad markers
bad_markers = ["", " ", "NA", "N/A", "na", "None", "-", "--", "NULL", "null", "nan"]
df.replace(bad_markers, np.nan, inplace=True)

# Clean numeric-looking columns that may be stored as object
def clean_numeric_col(col):
    if col in df.columns and df[col].dtype == object:
        # remove commas, currency symbols, percent signs, then coerce
        df[col] = (df[col].astype(str)
                          .str.replace(',', '')
                          .str.replace('%', '')
                          .str.replace('', '')
                          .str.strip())
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Apply to columns in likely list that sound numeric
for col in ["Biodegradation Time (days)", "Recyclability %", "Waste Reduction Impact %", "Carbon Footprint"]:
    if col in df.columns:
        clean_numeric_col(col)

# Recompute missing summary
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
from sklearn.impute import SimpleImputer, KNNImputer

IMPUTE_METHOD = "median"   # choose: "median", "mean", or "knn"

# Numeric columns to impute (restrict to the ones that need it or use all numeric)
numeric_for_impute = [c for c in num_cols if df[c].isna().sum()>0]
cat_for_impute = [c for c in cat_cols if df[c].isna().sum()>0]

print("Numeric needing impute:", numeric_for_impute)
print("Categorical needing impute:", cat_for_impute)

if IMPUTE_METHOD in ("median","mean"):
    num_strategy = "median" if IMPUTE_METHOD=="median" else "mean"
    num_imp = SimpleImputer(strategy=num_strategy)
    if numeric_for_impute:
        df[numeric_for_impute] = num_imp.fit_transform(df[numeric_for_impute])
    cat_imp = SimpleImputer(strategy="most_frequent", fill_value=None)
    if cat_for_impute:
        df[cat_for_impute] = cat_imp.fit_transform(df[cat_for_impute])

elif IMPUTE_METHOD=="knn":
    # KNNImputer operates on numeric arrays only  so we impute numeric with KNN and categorical with mode
    knn_imp = KNNImputer(n_neighbors=5)
    if numeric_for_impute:
        df[numeric_for_impute] = knn_imp.fit_transform(df[numeric_for_impute])
    cat_imp = SimpleImputer(strategy="most_frequent")
    if cat_for_impute:
        df[cat_for_impute] = cat_imp.fit_transform(df[cat_for_impute])

else:
    raise ValueError("Unsupported IMPUTE_METHOD")
```

```python
# Define critical columns  if a row misses many of these, drop it
critical_cols = ["Biodegradation Time (days)", "Recyclability %", "Carbon Footprint"]  # adjust as needed
critical_present = [c for c in critical_cols if c in df.columns]
max_missing_allowed = 2   # drop rows missing >= this many criticals

if critical_present:
    mask = df[critical_present].isnull().sum(axis=1) >= max_missing_allowed
    dropped = mask.sum()
    print(f"Dropping {dropped} rows with >= {max_missing_allowed} critical missing values")
    df = df[~mask].copy()

print("After handling critical drops, shape:", df.shape)
```

```python
remaining_missing = df.isnull().sum().sort_values(ascending=False)
remaining_missing = remaining_missing[remaining_missing>0]
print("Remaining missing after imputation/dropping:\n", remaining_missing)

# Save missing report
with open(os.path.join(DOCS_DIR, "missing_value_report.md"), "w") as f:
    f.write("# Missing Value Report\n\n")
    f.write(f"Imputation method: {IMPUTE_METHOD}\n\n")
    f.write("Remaining missing (post-processing):\n\n")
    f.write(remaining_missing.to_string())

# Save cleaned dataset
cleaned_fp = os.path.join(PROCESSED_DIR, "cleaned_integrated_materials.csv")
df.to_csv(cleaned_fp, index=False)
print("Saved cleaned dataset to:", cleaned_fp)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib

# Example  adjust these column names based on your df
nominal_cols = [c for c in ["Packaging Type", "Material Type", "Recyclability Category", "Supplier Region", "Recommended Use Cases"] if c in df.columns]
ordered_cols = []   # e.g., ["Quality Score"] if present and ordered
multilabel_cols = [c for c in ["Suitable Product Categories"] if c in df.columns]  # treat specially
numeric_cols = [c for c in df.select_dtypes(include=[np.number]).columns.tolist()]

print("Nominal:", nominal_cols)
print("Multilabel:", multilabel_cols)
print("Numeric:", numeric_cols)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

if multilabel_cols:
    ml_col = multilabel_cols[0]
    # assume separator is ';' or '|'  try both fallback
    def split_labels(x):
        if pd.isna(x): return []
        if ';' in x: return [s.strip() for s in x.split(';') if s.strip()]
        if '|' in x: return [s.strip() for s in x.split('|') if s.strip()]
        return [s.strip() for s in str(x).split(',') if s.strip()]

    mlb = MultiLabelBinarizer()
    mlb_matrix = mlb.fit_transform(df[ml_col].apply(split_labels))
    mlb_cols = [f"{ml_col}__{c}" for c in mlb.classes_]
    df_mlb = pd.DataFrame(mlb_matrix, columns=mlb_cols, index=df.index)
    # join and drop original column
    df = pd.concat([df.drop(columns=[ml_col]), df_mlb], axis=1)

    # save the mlb for inference
    joblib.dump(mlb, os.path.join(MODELS_DIR, "encoders", f"{ml_col}_mlb.pkl"))
    print("Saved MultiLabelBinarizer for", ml_col)
```

```python
from sklearn.compose import make_column_transformer

SCALER = "minmax"  # choose "minmax" or "standard"

# Recompute lists because df may have new mlb columns
all_numeric_cols = [c for c in df.select_dtypes(include=[np.number]).columns.tolist()]
all_cat_cols = [c for c in df.select_dtypes(include=['object','category']).columns.tolist()]

# Use only the originally detected nominal columns (intersect)
nominal_cols = [c for c in nominal_cols if c in all_cat_cols]

print("Final numeric cols:", all_numeric_cols)
print("Final categorical cols (to OHE):", nominal_cols)

# encoders/scalers
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
if SCALER=="minmax":
    scaler = MinMaxScaler()
else:
    scaler = StandardScaler()

preprocessor = ColumnTransformer(
    transformers=[
        ("num", scaler, all_numeric_cols),
        ("ohe", ohe, nominal_cols)
    ],
    remainder='passthrough'  # leave any other columns (e.g., already binary mlb cols)
)
```

```python
# Fit & transform
X_transformed = preprocessor.fit_transform(df)

# build feature names
# numeric scaled names
num_scaled_names = all_numeric_cols

# ohe names (requires get_feature_names_out)
ohe_feature_names = []
if nominal_cols:
    ohe_feature_names = preprocessor.named_transformers_['ohe'].get_feature_names_out(nominal_cols).tolist()

# remainder columns (those not transformed)  ColumnTransformer with remainder='passthrough' appends them at end in this order:
# we need to get names for passthrough columns
passthrough_cols = [c for c in df.columns if c not in all_numeric_cols + nominal_cols]

all_feature_names = num_scaled_names + ohe_feature_names + passthrough_cols

# Create DataFrame
X_encoded = pd.DataFrame(X_transformed, columns=all_feature_names, index=df.index)

print("Encoded dataframe shape:", X_encoded.shape)
X_encoded.head(3)
```

```python
# Save objects
joblib.dump(preprocessor.named_transformers_['ohe'], os.path.join(MODELS_DIR, "encoders", "onehot_encoder.pkl"))
joblib.dump(preprocessor.named_transformers_['num'], os.path.join(MODELS_DIR, "scalers", f"{SCALER}_scaler.pkl"))
joblib.dump(preprocessor, os.path.join(MODELS_DIR, "encoders", "column_transformer_pipeline.pkl"))

print("Saved encoders and scalers to", os.path.join(MODELS_DIR))
```

```python
# --- Configuration & Column names: update if your dataset uses different names ---
# numeric columns assumed in df:
col_co2 = "CO2_per_kg"                 # CO2 emissions per kg (numeric)
col_biodeg = "Biodegradation Time (days)"  # numeric days
col_recyclability = "Recyclability Category"  # A/B/C/D or similar
col_material_type = "Material Type"     # e.g., plastic, paper, metal, bio-based
col_cost_per_kg = "Cost_per_kg"         # cost numeric
col_weight_per_unit_g = "Weight_per_unit_g"  # weight grams (or kg if labeled)
col_durability = "Durability"          # numeric 0-10 or similar
# product requirement columns (optional) - if not present the MSS will use attribute averages
col_load = "Load_handling"
col_moisture = "Moisture_resistance"
col_thermal = "Thermal_resistance"

# Ensure columns exist - print missing ones
expected = [col_co2, col_biodeg, col_recyclability, col_material_type, col_cost_per_kg, col_weight_per_unit_g, col_durability]
print("Columns present:", [c for c in expected if c in df.columns])
```

```python
# helper functions
def safe_minmax_series(s):
    """Return min,max, but handle constant column by returning (min, min+1)."""
    mn = s.min(skipna=True)
    mx = s.max(skipna=True)
    if pd.isna(mn): return (0.0,1.0)
    if mn==mx:
        return (mn, mn+1.0)
    return (mn,mx)

def normalize_series(s, mn=None, mx=None):
    if mn is None or mx is None:
        mn, mx = safe_minmax_series(s)
    return (s - mn) / (mx - mn)

# 1) CO2 Impact Index (CII)
def compute_CII(df_in,
                co2_col=col_co2,
                bio_col=col_biodeg,
                rec_col=col_recyclability,
                mtype_col=col_material_type,
                weights = {"co2":0.5, "bio":0.25, "recy":0.2, "mtype":0.05}):
    df = df_in.copy()
    # map recyclability to 0..1 (0 best)
    rec_map = {"A":0.0, "B":0.33, "C":0.66, "D":1.0}
    df["_recy_score"] = df[rec_col].map(rec_map).fillna(0.5)  # default medium if missing

    # co2 normalize 0..1 (higher is worse)
    if co2_col in df.columns:
        co2_mn, co2_mx = safe_minmax_series(df[co2_col].dropna())
        df["_co2_norm"] = normalize_series(df[co2_col], co2_mn, co2_mx).fillna(df[co2_col].median() if df[co2_col].notna().any() else 0.5)
    else:
        df["_co2_norm"] = 0.5

    # biodegradation -> we consider longer days worse for environment (higher impact), so normalize days then invert for greenness
    if bio_col in df.columns:
        # use log scaling to soften very large outliers
        s = df[bio_col].replace(0, np.nan)  # avoid log(0)
        mn, mx = safe_minmax_series(np.log1p(s.dropna()))  # on log scale
        bio_log = np.log1p(df[bio_col].fillna(df[bio_col].median() if df[bio_col].notna().any() else 1))
        bio_norm = normalize_series(bio_log, mn, mx)
        # longer biodeg days -> higher normalized -> higher impact, so keep as-is
        df["_bio_norm"] = bio_norm.fillna(0.5)
    else:
        df["_bio_norm"] = 0.5

    # material type score: small domain bias (optional). Map types to multipliers 0..1 (0 best)
    type_map = {
        "bio-based": 0.0,
        "paper": 0.2,
        "metal": 0.3,
        "glass": 0.2,
        "plastic": 0.7,
        "composite": 0.6
    }
    df["_mtype_score"] = df[mtype_col].map(lambda x: type_map.get(str(x).lower(), 0.4) if pd.notna(x) else 0.4)

    # Weighted sum -> raw
    df["_cii_raw"] = (weights["co2"] * df["_co2_norm"].astype(float)
                     + weights["bio"] * df["_bio_norm"].astype(float)
                     + weights["recy"] * df["_recy_score"].astype(float)
                     + weights["mtype"] * df["_mtype_score"].astype(float))

    # scale to 0..100 (higher worse). normalize by max observed raw (or by theoretical max 1.0)
    max_possible = 1.0  # since components are normalized to 0..1 and weights sum to 1.0 by default
    df["CO2_Impact_Index_raw"] = df["_cii_raw"]
    df["CO2_Impact_Index"] = (df["_cii_raw"] / max_possible) * 100
    df["CO2_Impact_Index"] = df["CO2_Impact_Index"].clip(0,100).round(2)
    return df.drop(columns=["_recy_score","_co2_norm","_bio_norm","_mtype_score","_cii_raw"])

# 2) Cost Efficiency Index (CEI)
def compute_CEI(df_in,
                cost_col=col_cost_per_kg,
                weight_g_col=col_weight_per_unit_g,
                rec_col=col_recyclability,
                durability_col=col_durability,
                weights={"cost":0.6, "dur":0.3, "recy":0.1}):
    df = df_in.copy()
    # convert weight to kg if provided in grams
    if weight_g_col in df.columns:
        # assume column name implies grams; make new column weight_kg
        df["_weight_kg"] = df[weight_g_col] / 1000.0
    else:
        df["_weight_kg"] = 1.0  # fallback

    # cost per unit packaging
    if cost_col in df.columns:
        df["_cost_per_unit"] = df[cost_col] * df["_weight_kg"]
    else:
        df["_cost_per_unit"] = df.get("_cost_per_unit", df.get("cost_per_unit", np.nan)).fillna(1.0)

    # normalize cost (higher cost -> worse -> invert)
    mn, mx = safe_minmax_series(df["_cost_per_unit"].dropna())
    df["_cost_norm"] = normalize_series(df["_cost_per_unit"], mn, mx)
    df["_cost_eff"] = 1 - df["_cost_norm"]  # higher is better

    # durability normalize (assume scale 0-10 or detect)
    if durability_col in df.columns:
        dur_mn, dur_mx = safe_minmax_series(df[durability_col].dropna())
        df["_dur_norm"] = normalize_series(df[durability_col], dur_mn, dur_mx)
    else:
        df["_dur_norm"] = 0.5

    # recyclability bonus: A->+0.1, B->+0.06, C->+0.03, D->0
    rec_bonus_map = {"A":0.1, "B":0.06, "C":0.03, "D":0.0}
    df["_recy_bonus"] = df[rec_col].map(rec_bonus_map).fillna(0.02)

    # combine (bounded 0..1)
    df["_cei_raw"] = (weights["cost"] * df["_cost_eff"].astype(float)
                     + weights["dur"] * df["_dur_norm"].astype(float)
                     + weights["recy"] * df["_recy_bonus"].astype(float))

    # scale 0..100
    df["Cost_Efficiency_Index_raw"] = df["_cei_raw"]
    # since rec bonus can push >1 slightly, cap at 1
    df["Cost_Efficiency_Index"] = (df["_cei_raw"].clip(0,1) / 1.0) * 100
    df["Cost_Efficiency_Index"] = df["Cost_Efficiency_Index"].round(2)
    return df.drop(columns=["_weight_kg","_cost_per_unit","_cost_norm","_cost_eff","_dur_norm","_recy_bonus","_cei_raw"])

# 3) Material Suitability Score (MSS)
# This implementation assumes product requirements are columns in the dataset or we apply general matching.
def compute_MSS(df_in,
                load_col=col_load, moisture_col=col_moisture, thermal_col=col_thermal,
                durability_col=col_durability,
                weights={"load":0.3, "moisture":0.25, "thermal":0.25, "dur":0.2},
                mandatory_thresholds=None):
    df = df_in.copy()
    # if product requirement columns are missing, try to use material attribute columns or produce a general suitability based on normalized attributes
    # normalize each attribute
    for c in [load_col, moisture_col, thermal_col, durability_col]:
        if c in df.columns:
            mn, mx = safe_minmax_series(df[c].dropna())
            df[f"_{c}_norm"] = normalize_series(df[c], mn, mx).fillna(0.5)
        else:
            df[f"_{c}_norm"] = 0.5

    # Apply mandatory thresholds: if supplied as dict {column: threshold_value}, penalize heavy if below
    penalty = pd.Series(0, index=df.index, dtype=float)
    if mandatory_thresholds:
        for c, thr in mandatory_thresholds.items():
            if c in df.columns:
                below = df[c] < thr
                penalty.loc[below] += 0.5  # heavy penalty; you can tune
    # weighted sum
    df["_mss_raw"] = (weights["load"] * df[f"_{load_col}_norm"]
                    + weights["moisture"] * df[f"_{moisture_col}_norm"]
                    + weights["thermal"] * df[f"_{thermal_col}_norm"]
                    + weights["dur"] * df[f"_{durability_col}_norm"]) - penalty
    # clip 0..1 then scale 0..100
    df["Material_Suitability_Score_raw"] = df["_mss_raw"]
    df["Material_Suitability_Score"] = (df["_mss_raw"].clip(0,1) * 100).round(2)
    return df.drop(columns=[f"_{load_col}_norm", f"_{moisture_col}_norm", f"_{thermal_col}_norm", f"_{durability_col}_norm", "_mss_raw"])
```

```python
import pandas as pd, os
RAW = "/content/EcoPackAI_dataset.csv"
CLEANED_FP = "/content/ecopackai/data/processed/cleaned_integrated_materials.csv"
ENGINEERED_FP = "/content/ecopackai/data/model_ready/materials_engineered.csv"

# choose which to work from: prefer cleaned if exists else raw
if os.path.exists(CLEANED_FP):
    df = pd.read_csv(CLEANED_FP)
    print("Loaded cleaned:", CLEANED_FP)
elif os.path.exists(ENGINEERED_FP):
    df = pd.read_parquet(ENGINEERED_FP) if ENGINEERED_FP.endswith(".parquet") else pd.read_csv(ENGINEERED_FP)
    print("Loaded engineered:", ENGINEERED_FP)
elif os.path.exists(RAW):
    df = pd.read_csv(RAW)
    print("Loaded raw:", RAW)
else:
    raise FileNotFoundError("No dataset found: place your CSV at /content/EcoPackAI_dataset.csv or cleaned path")

print("Shape:", df.shape)
df.head(2)
```

```python
rules_md = """# Data Quality Rules  EcoPackAI

## Mandatory columns
- MaterialID (unique), Material, Cost_per_kg, CO2_per_kg

## Missing value rules
- No nulls in mandatory fields
- Optional: Supplier Notes, Special Handling

## Value ranges
- Cost_per_kg > 0
- CO2_per_kg >= 0
- Moisture_resistance: 110
- Thermal_resistance: 110
- Biodegradation Time (days) >= 1

## Categorical valid sets
- Packaging Type: Box, Pouch, Tray, Wrap, Compostable Sheet
- Material Type: Paper, Plastic, Metal, Bio-based
- Recyclability Category: A, B, C, D

## Uniqueness & integrity
- Material ID must be unique
- No duplicate rows
- No negative weights or dimensions

"""
with open("/content/ecopackai/docs/data_quality_rules.md","w") as f:
    f.write(rules_md)
print("Saved /content/ecopackai/docs/data_quality_rules.md")
```

```python
import numpy as np

dfc = df.copy()

# required columns: adapt names if yours differ
required_cols = ["MaterialID","Material","Cost_per_kg","CO2_per_kg"]
for c in required_cols:
    if c not in dfc.columns:
        print("WARNING: expected column not found:", c)

# numeric fill
num_cols = dfc.select_dtypes(include=['number']).columns.tolist()
for c in num_cols:
    median = dfc[c].median()
    dfc[c] = dfc[c].fillna(median)

# categorical fill
cat_cols = dfc.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    dfc[c] = dfc[c].fillna("Unknown")

# remove exact duplicates
before = dfc.shape[0]
dfc = dfc.drop_duplicates()
after = dfc.shape[0]

# ensure non-negative for some numeric fields
for c in ["Cost_per_kg","CO2_per_kg","Weight_per_unit_g"]:
    if c in dfc.columns:
        dfc[c] = dfc[c].abs()  # convert negatives to positive; if you prefer, set to NaN and re-impute

# save cleaned
os.makedirs(os.path.dirname(CLEANED_FP), exist_ok=True)
dfc.to_csv(CLEANED_FP, index=False)
print("Saved cleaned dataset:", CLEANED_FP, " (rows removed dupes:", before-after,")")
```

```python
test_code = r'''
import pandas as pd
import numpy as np

DF_PATH = "/content/ecopackai/data/processed/cleaned_integrated_materials.csv"

def load_df():
    return pd.read_csv(DF_PATH)

def test_required_columns_present():
    df = load_df()
    required = ["Material ID","Material Type","Cost per Unit (USD)","CO2 Emission per kg (estimated)"] # Adjusted column names
    for c in required:
        assert c in df.columns, f"Missing required column: {c}"

def test_no_nan_in_mandatory():
    df = load_df()
    # Adjusted mandatory column names to match the DataFrame
    mandatory = ["Material ID", "Packaging Type", "Material Type", "Cost per Unit (USD)", "CO2 Emission per kg (estimated)"]
    for c in mandatory:
        assert df[c].isna().sum() == 0, f"NaNs in mandatory col: {c}"

def test_unique_materialid():
    df = load_df()
    assert df["Material ID"].is_unique, "Material ID not unique" # Adjusted column name

def test_no_duplicates_rows():
    df = load_df()
    assert df.duplicated().sum() == 0, "There are duplicate rows"

def test_value_ranges():
    df = load_df()
    # Adjusted column names to match the DataFrame
    if "Cost per Unit (USD)" in df.columns:
        assert (df["Cost per Unit (USD)"] > 0).all(), "Cost per Unit (USD) must be > 0"
    if "CO2 Emission per kg (estimated)" in df.columns:
        assert (df["CO2 Emission per kg (estimated)"] >= 0).all(), "CO2 Emission per kg (estimated) must be >= 0"
    if "Biodegradation Time (days)" in df.columns:
        assert (df["Biodegradation Time (days)"] >= 1).all(), "Biodeg days >= 1"
    # Adjusted column names to match the DataFrame
    for c in ["Moisture Resistance Score","Thermal Resistance Score"]:
        if c in df.columns:
            assert df[c].between(1,10).all(), f"{c} must be 1..10"

def test_categorical_values():
    df = load_df()
    # example checks:
    if "Recyclability Category" in df.columns:
        assert set(df["Recyclability Category"].dropna().unique()).issubset(set(["High","Medium","Unknown"])), "Invalid recyclability categories" # Adjusted expected categories

def test_feature_columns_range_if_present():
    df = load_df()
    for feat in ["CO2_Impact_Index","Cost_Efficiency_Index","Material_Suitability_Score"]:
        if feat in df.columns:
            assert df[feat].between(0,100).all(), f"{feat} must be 0..100"
'''
open("/content/ecopackai/tests/test_data_quality.py","w").write(test_code)
print("Wrote /content/ecopackai/tests/test_data_quality.py")
```

```python
df_sample = pd.read_csv(CLEANED_FP)  # cleaned dataset
cols = []
for c in df_sample.columns:
    dtype = str(df_sample[c].dtype)
    example = df_sample[c].dropna().astype(str).iloc[0] if df_sample[c].dropna().shape[0]>0 else ""
    nullable = df_sample[c].isna().any()
    rng = ""
    if pd.api.types.is_numeric_dtype(df_sample[c]):
        rng = f"{df_sample[c].min()} .. {df_sample[c].max()}"
    elif pd.api.types.is_object_dtype(df_sample[c]):
        vals = df_sample[c].dropna().unique().tolist()
        if len(vals)<=10:
            rng = ", ".join(map(str,vals))
    cols.append({"column":c,"dtype":dtype,"nullable":nullable,"range_or_values":rng,"example":example,"derived":"No","used_in_ML":"Yes"})

dd = pd.DataFrame(cols)
md_path = os.path.join(BASE,"docs","data_dictionary_v2.md")
xlsx_path = os.path.join(BASE,"docs","data_dictionary_v2.xlsx")

# Save Markdown
with open(md_path,"w") as f:
    f.write("# Data Dictionary v2\n\n")
    f.write(dd.to_markdown(index=False))
# Save Excel
dd.to_excel(xlsx_path, index=False)
print("Saved data dictionary MD and XLSX to", BASE+"/docs/")
```
